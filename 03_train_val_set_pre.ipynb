```python
#Training Set    
import shutil
import os
# import glob 


source_file_dir = 'C:/Users/4Kims/0_python/x_ray/data/labels/'
destination_dir = 'C:/Users/4Kims/0_python/x_ray/data/labels/train/'

# label_files = glob.glob('C:/Users/User/obj_detect_ai_hub/labels/*.txt')

label_path_dir = 'C:/Users/4Kims/0_python/x_ray/data/labels'
label_file_list = os.listdir(label_path_dir)


img_path_dir = 'C:/Users/4Kims/0_python/x_ray/data/images/train/'
img_file_list = os.listdir(img_path_dir)
# i = 0

for label_file in label_file_list:
    label_file_name = label_file.split(".")
    print(label_file_name[0])
    
    for img_file in img_file_list:
        img_file_name = img_file.split(".")
        if label_file_name[0] == img_file_name[0]:
#             i += 1
            source_file = source_file_dir + label_file_name[0] + ".txt"
            destination_file = destination_dir + label_file_name[0] + ".txt"
            shutil.move(source_file,destination_file) 
            break
```
